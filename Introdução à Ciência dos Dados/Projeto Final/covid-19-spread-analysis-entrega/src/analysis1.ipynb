```python
#Tamanho da população de países Sul-Americanos
pop_ARG = 45195774
pop_BOL = 11673021
pop_BRA = 212559426 
pop_CHI = 19116201
pop_COL = 50882891
pop_ECU = 17643054
pop_GUY = 786552
pop_PAR = 7132538
pop_PER = 33050325
pop_SUR = 588552
pop_URU = 3473730
pop_VEN = 28435940

#Separando os paises com mais de 20 milhoes de pessoas e paises com menos de 20 milhoes de pessoas

#Total de Casos em 30 dias de pandemia
Values_30_UP = np.zeros(4)
Values_30_UP[0] = df_ARG.iloc[30]['total_cases']/pop_ARG 
Values_30_UP[1] = df_COL.iloc[30]['total_cases']/pop_COL
Values_30_UP[2] = df_PER.iloc[30]['total_cases']/pop_PER
Values_30_UP[3] = df_VEN.iloc[30]['total_cases']/pop_VEN
Values_30_UP = Values_30_UP * 100

Values_30_DOWN = np.zeros(7)
Values_30_DOWN[0] = df_BOL.iloc[30]['total_cases']/pop_BOL
Values_30_DOWN[1] = df_CHI.iloc[30]['total_cases']/pop_CHI 
Values_30_DOWN[2] = df_ECU.iloc[30]['total_cases']/pop_ECU
Values_30_DOWN[3] = df_GUY.iloc[30]['total_cases']/pop_GUY 
Values_30_DOWN[4] = df_PAR.iloc[30]['total_cases']/pop_PAR  
Values_30_DOWN[5] = df_SUR.iloc[30]['total_cases']/pop_SUR 
Values_30_DOWN[6] = df_URU.iloc[30]['total_cases']/pop_URU
Values_30_DOWN = Values_30_DOWN * 100

#Total de Casos em 60 dias de pandemia
Values_60_UP = np.zeros(4)
Values_60_UP[0] = df_ARG.iloc[60]['total_cases']/pop_ARG 
Values_60_UP[1] = df_COL.iloc[60]['total_cases']/pop_COL
Values_60_UP[2] = df_PER.iloc[60]['total_cases']/pop_PER 
Values_60_UP[3] = df_VEN.iloc[60]['total_cases']/pop_VEN
Values_60_UP = Values_60_UP * 100

Values_60_DOWN = np.zeros(7)
Values_60_DOWN[0] = df_BOL.iloc[60]['total_cases']/pop_BOL
Values_60_DOWN[1] = df_CHI.iloc[60]['total_cases']/pop_CHI 
Values_60_DOWN[2] = df_ECU.iloc[60]['total_cases']/pop_ECU
Values_60_DOWN[3] = df_GUY.iloc[60]['total_cases']/pop_GUY 
Values_60_DOWN[4] = df_PAR.iloc[60]['total_cases']/pop_PAR 
Values_60_DOWN[5] = df_SUR.iloc[60]['total_cases']/pop_SUR 
Values_60_DOWN[6] = df_URU.iloc[60]['total_cases']/pop_URU
Values_60_DOWN = Values_60_DOWN * 100

#Total de Casos em 90 dias de pandemia
Values_90_UP = np.zeros(4)
Values_90_UP[0] = df_ARG.iloc[90]['total_cases']/pop_ARG
Values_90_UP[1] = df_COL.iloc[90]['total_cases']/pop_COL
Values_90_UP[2] = df_PER.iloc[90]['total_cases']/pop_PER 
Values_90_UP[3] = df_VEN.iloc[90]['total_cases']/pop_VEN
Values_90_UP = Values_90_UP * 100

Values_90_DOWN = np.zeros(7)
Values_90_DOWN[0] = df_BOL.iloc[90]['total_cases']/pop_BOL
Values_90_DOWN[1] = df_CHI.iloc[90]['total_cases']/pop_CHI 
Values_90_DOWN[2] = df_ECU.iloc[90]['total_cases']/pop_ECU
Values_90_DOWN[3] = df_GUY.iloc[90]['total_cases']/pop_GUY 
Values_90_DOWN[4] = df_PAR.iloc[90]['total_cases']/pop_PAR 
Values_90_DOWN[5] = df_SUR.iloc[90]['total_cases']/pop_SUR 
Values_90_DOWN[6] = df_URU.iloc[90]['total_cases']/pop_URU
Values_90_DOWN = Values_90_DOWN * 100
```

```python
#Função de bootstrap para realizar os testes
def bootstrap_mean(x, n=5000):
    size = len(x)
    values = np.zeros(n)
    for i in range(n):
        sample = np.random.choice(x, size=size, replace=True)
        values[i] = sample.mean()
    return values

#Realizando os testes
BT_30_UP = bootstrap_mean(Values_30_UP)
BT_30_DOWN = bootstrap_mean(Values_30_DOWN)
IC_30_UP = [np.percentile(BT_30_UP, 2.5), np.percentile(BT_30_UP, 97.5)]
IC_30_DOWN = [np.percentile(BT_30_DOWN, 2.5), np.percentile(BT_30_DOWN, 97.5)]

BT_60_UP = bootstrap_mean(Values_60_UP)
BT_60_DOWN = bootstrap_mean(Values_60_DOWN)
IC_60_UP = [np.percentile(BT_60_UP, 2.5), np.percentile(BT_60_UP, 97.5)]
IC_60_DOWN = [np.percentile(BT_60_DOWN, 2.5), np.percentile(BT_60_DOWN, 97.5)]

BT_90_UP = bootstrap_mean(Values_90_UP)
BT_90_DOWN = bootstrap_mean(Values_90_DOWN)
IC_90_UP = [np.percentile(BT_90_UP, 2.5), np.percentile(BT_90_UP, 97.5)]
IC_90_DOWN = [np.percentile(BT_90_DOWN, 2.5), np.percentile(BT_90_DOWN, 97.5)]
```

```python
print('IC de países com mais de 20 milhões de habitantes depois de 30 dias: {}'.format(IC_30_UP))
print('IC de países com menos de 20 milhões de habitantes depois de 30 dias: {}'.format(IC_30_DOWN))
fig, ax = plt.subplots(figsize=(12, 8))
ax.hist(BT_30_UP, bins=20, edgecolor='k', label='UP', density=True)
ax.hist(BT_30_DOWN, bins=20, edgecolor='k', label='DOWN', density=True)
plt.axvline(IC_30_UP[0], color='r', linestyle='dashed', linewidth=1.5)
plt.axvline(IC_30_UP[1], color='r', linestyle='dashed', linewidth=1.5)
plt.axvline(IC_30_DOWN[0], color='k', linestyle='dashed', linewidth=1.5)
plt.axvline(IC_30_DOWN[1], color='k', linestyle='dashed', linewidth=1.5)
plt.xlabel("Porcentagem da população contaminada")
plt.title("30 dias")
ax.legend()
```

Acima temos o histograma da porcentagem da população contaminada no 30º dia de pandemia, as barras azuis representam países que possuem mais de 20 milhões de habitantes e a barra laranja países com menos que 20 milhões de habitantes, ainda no histograma temos duas linhas pontilhadas distintas, a linha pontilhada vermelha representa o IC de paises com mais de 20 milhões de habitantes e a linha pontilhada preta representa o IC de paises com mais de 20 milhões de habitantes, é visível que elas se cruzam logo não podemos confirmar a hipotese nula.
